```python
def is_DamageHealerWorks(match_id):
    import requests
    resp = requests.get('https://api.opendota.com/api/matches/{}'.format(match_id))
    data = resp.json()
    win_camp = 'Radiant' if data['radiant_win'] else 'Dire'
    for objective in data['objectives']:
        if 'key' in objective:
            if 'healers' in str(objective['key']):
                healer_camp = 'Dire' if 'goodguys' in str(objective['key']) else 'Radiant'
                #print(objective['time'],objective['key'])
                if win_camp == healer_camp:
                    return True
                else:
                    return False
```
